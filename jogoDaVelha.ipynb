```python
#VARIÁVEIS GLOBAIS, DEFININDO LISTA E QUANTIDADE DE ESCOLHA
lista=["_"]*9
qtd_escolha=0
```

```python
#FUNÇÃO PARA IMPRIMIR LISTA
def imprime_lista(lista):
  for i in range(len(lista)):
    print(lista[i], end=" ")
    if i==2 or i==5 or i==8:
      print("")
```

```python
#FUNÇÃO PARA VERIFICAR LISTA
def verifica_lista(lista,player):
  #valor escolhido
  def valor():
    if player=="X":
      return 1
    else:
      return 2
  #testes linhas   
   if lista[0] ==lista[1]==lista[2] and lista[0]==player:
    return valor()
  if lista[3] ==lista[4]==lista[5] and lista[3]==player:
    return valor()
  if lista[6] ==lista[7]==lista[8] and lista[6]==player:
    return valor() 

  #testes de colunas  
  if lista[0]==lista[3]==lista[6] and lista[0]==player:
    return valor()  
  if lista[1]==lista[4]==lista[7] and lista[1]==player:
    return valor()  
  if lista[2]==lista[5]==lista[8] and lista[2]==player:
    return valor()

  #testes linhas diagonais  
  if lista[0]==lista[4]==lista[8] and lista[0]==player:
    return valor() 
  if lista[2]==lista[4]==lista[6] and lista[2]==player:
    return valor() 
  return 0      
```

```python
#LAÇO DE REPETIÇÃO JOGO DA VELHA
while True:
  #escolha do usuário
  escolha= int(input("Qual é a sua escolha? "));
  if escolha >9 or escolha<0:
    print("Escolha inválida! Tente novamente\n")
    escolha= int(input("Qual é a sua escolha? ")); 

  while lista[escolha-1]!="_":
    print("Escolha inválida! Tente novamente\n")
    imprime_lista(lista)
    escolha= int(input("Qual é a sua escolha? "));

  lista[escolha-1]="X"
  qtd_escolha+=1

  ganhador=verifica_lista(lista,"X")

  if ganhador!=0 or qtd_escolha==9:
    break
  print("O status da lista é:\n ")
  imprime_lista(lista)
  print("\n")

 #escolha do computador
  escolha_PC=random.randint(1,9)

  while lista[escolha_PC-1] !="_":
    escolha_PC=random.randint(1,9)

  lista[escolha_PC-1]="O"
  qtd_escolha+=1

  ganhador=verifica_lista(lista,"O")
  
  if ganhador!=0:
    break
  print("O status da lista é:\n ")  
  imprime_lista(lista)
  print("\n")

#Verificar ganhador
if ganhador ==1:
  print("Parabéns, você ganhou!!\n")
elif ganhador ==2:
  print("Você perdeu, o PC ganhou!!\n")
else:
  print("Deu velha!!\n")
imprime_lista(lista)
 
```
